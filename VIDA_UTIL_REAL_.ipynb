```python

# Instalar dependências
!pip install scikit-learn pandas

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from datetime import datetime, timedelta

# ============================
# Entrada de dados do usuário
# ============================
Cinzas_t = float(input("Informe Cinzas totais (t): "))
Cinzas_t_h = float(input("Informe Cinzas t/h: "))
Torque = float(input("Informe Torque (Nm): "))

# ============================
# Fórmula original
# ============================
vida_util_formula = 2091.7335 * (Cinzas_t ** -0.3945) * (Cinzas_t_h ** -0.2226) * (Torque ** -0.1726)

# ============================
# Gerar dados sintéticos para treino
# ============================
np.random.seed(42)
X = np.random.rand(500, 3) * [20, 15, 300]  # Cinzas_t, Cinzas_t_h, Torque
Y = 2091.7335 * (X[:, 0] ** -0.3945) * (X[:, 1] ** -0.2226) * (X[:, 2] ** -0.1726)

# ============================
# Treinar modelos
# ============================
ml_model = LinearRegression()
ml_model.fit(X, Y)
pred_ml = ml_model.predict([[Cinzas_t, Cinzas_t_h, Torque]])[0]

nn_model = MLPRegressor(hidden_layer_sizes=(5,), max_iter=1000, random_state=42)
nn_model.fit(X, Y)
pred_nn = nn_model.predict([[Cinzas_t, Cinzas_t_h, Torque]])[0]

dl_model = MLPRegressor(hidden_layer_sizes=(50, 30, 10), max_iter=2000, random_state=42)
dl_model.fit(X, Y)
pred_dl = dl_model.predict([[Cinzas_t, Cinzas_t_h, Torque]])[0]

# ============================
# Calcular datas previstas
# ============================
hoje = datetime.now()
data_formula = hoje + timedelta(hours=vida_util_formula)
data_ml = hoje + timedelta(hours=pred_ml)
data_nn = hoje + timedelta(hours=pred_nn)
data_dl = hoje + timedelta(hours=pred_dl)

# ============================
# Exibir resultados
# ============================
print("\n=== RESULTADOS ===")
print(f"Dados fornecidos: Cinzas_t = {Cinzas_t}, Cinzas_t/h = {Cinzas_t_h}, Torque = {Torque}")
print("\nPrevisões de vida útil (em horas):")
print(f"Fórmula: {vida_util_formula:.2f} h")
print(f"Machine Learning (Regressão Linear): {pred_ml:.2f} h")
print(f"Rede Neural Simples: {pred_nn:.2f} h")
print(f"Deep Learning: {pred_dl:.2f} h")

print("\nDatas previstas de falha:")
print(f"Fórmula: {data_formula}")
print(f"Machine Learning: {data_ml}")
print(f"Rede Neural Simples: {data_nn}")
print(f"Deep Learning: {data_dl}")

# ============================
# Salvar resultados em CSV
# ============================
dados = {
    "Cinzas_t": [Cinzas_t],
    "Cinzas_t_h": [Cinzas_t_h],
    "Torque": [Torque],
    "Vida_Formula_h": [vida_util_formula],
    "Vida_ML_h": [pred_ml],
    "Vida_NN_h": [pred_nn],
    "Vida_DL_h": [pred_dl],
    "Data_Formula": [data_formula],
    "Data_ML": [data_ml],
    "Data_NN": [data_nn],
    "Data_DL": [data_dl]
}

df = pd.DataFrame(dados)
df.to_csv("previsao_vida_util.csv", index=False)
```
